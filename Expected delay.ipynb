```python
# Getting delays of the train for past week

rows = [row for row in soup.find("table").find_all("tr")]
#print rows
#print type(rows[0])
size = len(rows)-1
delays = [rows[i].find_all("td")[2] for i in range(1,size)]
delays = [delay.contents[0] for delay in delays]
print delays
#delays = [delay[2] for delay in rows.find_all("td")]
```

```python
# Converting the time into minutes
#print delays[0][0:2]
avgdelay=0;
for delay in delays:
    if delay[0] == '-':
        pass
    elif delay[3] == 'M':
        avgdelay=avgdelay+int(delay[0:2])
    else:
        avgdelay=avgdelay+int(delay[0:2])*60+int(delay[7:9])
print avgdelay/len(delays)
#delays = [ for delay in delays]
```
